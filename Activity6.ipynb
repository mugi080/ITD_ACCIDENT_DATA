```python

import folium
from folium.plugins import HeatMap
```

```python
df['country_txt'].unique()
```

```python
filter1 = df[df['country_txt'] == 'Philippines' ]
```

```python
filter1['attacktype1_txt'].unique()
```

```python
filter2 = filter1[filter1['attacktype1_txt'] == 'Hostage Taking (Kidnapping)']
```

```python
filter2
```

```python
hostage_ph = filter2.groupby(['latitude', 'longitude']).size().reset_index(name = 'incident_count')
```

```python
hostage_ph
```

```python
hostage_ph.max()
```

```python

#OPTIONAL, for clearer view for the Geospatial Map

hostage_ph["normalize_data"] = hostage_ph['incident_count'] - hostage_ph['incident_count'].min() / \
                                hostage_ph['incident_count'].max() - hostage_ph['incident_count']

#geospatial folium data should be formatted as a list

hostage_ph_list = hostage_ph[['latitude', 'longitude', 'incident_count']].values.tolist()
```

```python

my_map1 = folium.Map(location = [10, 120], zoom_start = 6)
```

```python

HeatMap(hostage_ph_list).add_to(my_map1)
```

```python

my_map1.save('regional_analysis.html')
print("data save to regional_analysis.html")
```

```python

my_map1
```
